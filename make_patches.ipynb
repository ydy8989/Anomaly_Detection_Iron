```python
import PIL.Image as pilimg # Pillow 설치 해야함 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

im = pilimg.open('/home/pirl/Downloads/phone/hwang/ano/20181128_135625.jpg')


pix = np.array(im)

np.shape(pix)
```

```python
plt.imshow(pix)
```

```python
from sklearn.feature_extraction import image

patches = image.extract_patches_2d(pix, (2000,2000), max_patches=2, random_state=100)
# (파일명, 원하는 크기, 원하는 패치 개수, 랜덤으로 뽑느냐(NONE으로 해도 됨))
```

```python
np.shape(patches)
```

```python
# 디렉토리에 이미지 저장 
for i in range(np.shape(patches)[0]):
    Image.fromarray(patches[i]).save('/home/pirl/Downloads/phone/hwang/anomaly_patch/%s_3.jpg'%i)
```
